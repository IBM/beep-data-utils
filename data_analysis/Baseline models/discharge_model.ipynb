```python
import matplotlib.pyplot as plt
import numpy as np
import torch
from scipy.interpolate import interp1d
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error


def interpolate(x, y, interp_dims, xs=0, xe=1):
    """Interpolate function from original implementation"""
    if len(x) <= 2:
        return np.zeros(interp_dims)
    func = interp1d(x, y, bounds_error=False)
    new_x = np.linspace(xs, xe, interp_dims)
    return func(new_x)


def get_Qdlin(cycle_data, min_V, max_V):
    """Calculate Qdlin matching the original implementation"""
    eps = 1e-1
    I = np.array(cycle_data["current"])
    V = np.array(cycle_data["voltage"])
    Q = np.array(cycle_data["discharge_capacity"])

    discharge_mask = I < -eps
    if sum(discharge_mask) <= 2:
        return np.zeros(1000)

    return interpolate(V[discharge_mask], Q[discharge_mask], 1000, xs=min_V, xe=max_V)[::-1]


def smooth(x, window_size=10, sigma=3):
    """Smooth function from original implementation"""
    res = np.empty_like(x)
    meds = np.empty_like(x)
    for i in range(len(x)):
        low = max(0, i - window_size)
        high = min(len(x), i + window_size + 1)
        meds[i] = np.median(x[low:high])
    base = np.std(np.abs(x - meds))
    for i in range(len(x)):
        if np.abs(meds[i] - x[i]) > base * sigma:
            res[i] = meds[i]
        else:
            res[i] = x[i]
    return meds


def get_discharge_features(battery_data, cycle_early=10, cycle_late=100):
    """Extract discharge features using PyTorch tensors like the original"""
    eps = 1e-8
    min_V = float(battery_data["voltage"].min())
    max_V = float(battery_data["voltage"].max())

    early_cycle = battery_data[battery_data["cycle_number"] == cycle_early]
    late_cycle = battery_data[battery_data["cycle_number"] == cycle_late]

    if len(early_cycle) < 2 or len(late_cycle) < 2:
        return None

    # Calculate Qdlin curves
    early_qdlin = get_Qdlin(early_cycle, min_V, max_V)
    late_qdlin = get_Qdlin(late_cycle, min_V, max_V)

    if np.all(early_qdlin == 0) or np.all(late_qdlin == 0):
        return None

    diff_qdlin = late_qdlin - early_qdlin
    diff_qdlin = smooth(diff_qdlin)  # Apply smoothing like original
    diff_qdlin = torch.from_numpy(diff_qdlin).float()

    diff_qdlin = diff_qdlin[~torch.isnan(diff_qdlin)]
    if len(diff_qdlin) <= 1:
        return None

    features = {}

    features["Minimum"] = (diff_qdlin.min().abs() + eps).log10().item()
    features["Variance"] = (diff_qdlin.var() + eps).log10().item()

    x_mean = diff_qdlin.mean()
    x_std = diff_qdlin.std()

    features["Skewness"] = (
        (((diff_qdlin - x_mean) ** 3).mean().abs() / (x_std**3 + eps) + eps).log10().item()
    )

    features["Kurtosis"] = (
        (((diff_qdlin - x_mean) ** 4).mean() / (diff_qdlin.var() ** 2 + eps) + eps).log10().item()
    )

    cycle_numbers = range(cycle_early, cycle_late + 1)
    cycle_data = battery_data[battery_data["cycle_number"].isin(cycle_numbers)]
    discharge_capacities = cycle_data.groupby("cycle_number")["discharge_capacity"].max()

    if len(discharge_capacities) > 0:
        features["Early discharge capacity"] = float(discharge_capacities.iloc[0])
        features["Difference between max discharge capacity and early discharge capacity"] = float(
            discharge_capacities.max() - discharge_capacities.iloc[0]
        )
    else:
        return None

    if not all(np.isfinite(val) for val in features.values()):
        return None

    return features


def evaluate_model(split_name, train_cells, test_cells, df):
    """Evaluate model on a train/test split"""
    print(f"\nProcessing split: {split_name}")

    X_train, y_train = [], []
    feature_names = [
        "Minimum",
        "Variance",
        "Skewness",
        "Kurtosis",
        "Early discharge capacity",
        "Difference between max discharge capacity and early discharge capacity",
    ]

    for cell in train_cells:
        cell_data = df[df["cell_key"] == cell]
        features = get_discharge_features(cell_data)
        if features is not None:
            X_train.append([features[name] for name in feature_names])
            y_train.append(cell_data["cycle_life"].iloc[0])
        else:
            print(f"Could not extract features for train cell {cell}")

    # Extract features for test cells
    X_test, y_test = [], []
    for cell in test_cells:
        cell_data = df[df["cell_key"] == cell]
        features = get_discharge_features(cell_data)
        if features is not None:
            X_test.append([features[name] for name in feature_names])
            y_test.append(cell_data["cycle_life"].iloc[0])
        else:
            print(f"Could not extract features for test cell {cell}")

    if len(X_train) == 0 or len(X_test) == 0:
        print("Not enough data for modeling")
        return None

    X_train = np.array(X_train)
    X_test = np.array(X_test)
    y_train = np.array(y_train)
    y_test = np.array(y_test)

    model = LinearRegression()
    model.fit(X_train, y_train)

    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)

    results = {
        "train_rmse": np.sqrt(mean_squared_error(y_train, y_pred_train)),
        "test_rmse": np.sqrt(mean_squared_error(y_test, y_pred_test)),
        "train_mape": mean_absolute_percentage_error(y_train, y_pred_train) * 100,
        "test_mape": mean_absolute_percentage_error(y_test, y_pred_test) * 100,
        "n_train": len(y_train),
        "n_test": len(y_test),
        "feature_importances": dict(zip(feature_names, model.coef_)),
    }

    # Print results
    print(f"\nResults for {split_name}:")
    print(f"Train RMSE: {results['train_rmse']:.0f} cycles")
    print(f"Test RMSE: {results['test_rmse']:.0f} cycles")
    print(f"Train MAPE: {results['train_mape']} cycles")
    print(f"Test MAPE: {results['test_mape']} cycles")
    print(f"Number of training cells: {results['n_train']}")
    print(f"Number of test cells: {results['n_test']}")

    # Plot results
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.scatter(y_train, y_pred_train, alpha=0.5, label="Train")
    plt.scatter(y_test, y_pred_test, alpha=0.5, label="Test")
    max_cycles = max(max(y_train), max(y_test))
    plt.plot([0, max_cycles], [0, max_cycles], "k--")
    plt.xlabel("True Cycle Life")
    plt.ylabel("Predicted Cycle Life")
    plt.legend()
    plt.title(f"{split_name}: Prediction vs Ground Truth")

    plt.subplot(1, 2, 2)
    plt.bar(feature_names, model.coef_)
    plt.xticks(rotation=45, ha="right")
    plt.title("Feature Importances")

    plt.tight_layout()
    plt.show()

    return results


def test_on_sample():
    main_df = pd.read_csv("../beep_main_ds_quartiles.csv")
    secondary_df = pd.read_csv("../test_secondary_beep_dataset_quartile_charge_policy.csv")
    df = pd.concat([main_df, secondary_df])
    # df = pd.read_csv("/U..nc/nca_ncm_cycle_life.csv")
    # df = df[df["cycle_life"] >= 100]
    # df = df[~df["cell_key"].str.startswith("ncm")]
    # df = df[~df["cell_key"].str.startswith("ncmca")]

    for k, v in get_cell_splits().items():
        print(
            "-" * 50 + "\n",
            f"{k}",
        )
        results = evaluate_model("Test Split", v["train"], v["test"], df)
        if results:
            print(f"\nFeature Importances for model {k}:")
            for feature, importance in results["feature_importances"].items():
                print(f"{feature}: {importance:.4f}")


if __name__ == "__main__":
    test_on_sample()
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler


def interpolate(x, y, interp_dims, xs=0, xe=1):
    """Interpolate function from original implementation"""
    if len(x) <= 2:
        return np.zeros(interp_dims)
    func = interp1d(x, y, bounds_error=False)
    new_x = np.linspace(xs, xe, interp_dims)
    return func(new_x)


def get_Qdlin(cycle_data, min_V, max_V):
    """Calculate Qdlin matching the original implementation"""
    eps = 1e-1
    I = np.array(cycle_data["current"])
    V = np.array(cycle_data["voltage"])
    Q = np.array(cycle_data["discharge_capacity"])

    discharge_mask = I < -eps
    if sum(discharge_mask) <= 2:
        return np.zeros(1000)

    return interpolate(V[discharge_mask], Q[discharge_mask], 1000, xs=min_V, xe=max_V)[::-1]


def smooth(x, window_size=10, sigma=3):
    """Smooth function from original implementation"""
    res = np.empty_like(x)
    meds = np.empty_like(x)
    for i in range(len(x)):
        low = max(0, i - window_size)
        high = min(len(x), i + window_size + 1)
        meds[i] = np.median(x[low:high])
    base = np.std(np.abs(x - meds))
    for i in range(len(x)):
        if np.abs(meds[i] - x[i]) > base * sigma:
            res[i] = meds[i]
        else:
            res[i] = x[i]
    return meds


def get_discharge_features(battery_data, cycle_early=10, cycle_late=100):
    """Extract discharge features using PyTorch tensors like the original"""
    eps = 1e-8
    # Get voltage limits
    min_V = float(battery_data["voltage"].min())
    max_V = float(battery_data["voltage"].max())

    # Get early and late cycles
    early_cycle = battery_data[battery_data["cycle_number"] == cycle_early]
    late_cycle = battery_data[battery_data["cycle_number"] == cycle_late]

    if len(early_cycle) < 2 or len(late_cycle) < 2:
        return None

    early_qdlin = get_Qdlin(early_cycle, min_V, max_V)
    late_qdlin = get_Qdlin(late_cycle, min_V, max_V)

    if np.all(early_qdlin == 0) or np.all(late_qdlin == 0):
        return None

    diff_qdlin = late_qdlin - early_qdlin
    diff_qdlin = smooth(diff_qdlin)
    diff_qdlin = torch.from_numpy(diff_qdlin).float()

    diff_qdlin = diff_qdlin[~torch.isnan(diff_qdlin)]
    if len(diff_qdlin) <= 1:
        return None

    features = {}

    features["Minimum"] = (diff_qdlin.min().abs() + eps).log10().item()
    features["Variance"] = (diff_qdlin.var() + eps).log10().item()

    x_mean = diff_qdlin.mean()
    x_std = diff_qdlin.std()

    features["Skewness"] = (
        (((diff_qdlin - x_mean) ** 3).mean().abs() / (x_std**3 + eps) + eps).log10().item()
    )

    features["Kurtosis"] = (
        (((diff_qdlin - x_mean) ** 4).mean() / (diff_qdlin.var() ** 2 + eps) + eps).log10().item()
    )

    cycle_numbers = range(cycle_early, cycle_late + 1)
    cycle_data = battery_data[battery_data["cycle_number"].isin(cycle_numbers)]
    discharge_capacities = cycle_data.groupby("cycle_number")["discharge_capacity"].max()

    if len(discharge_capacities) > 0:
        features["Early discharge capacity"] = float(discharge_capacities.iloc[0])
        features["Difference between max discharge capacity and early discharge capacity"] = float(
            discharge_capacities.max() - discharge_capacities.iloc[0]
        )
    else:
        return None

    if not all(np.isfinite(val) for val in features.values()):
        return None

    return features


def extract_features_for_visualization(df):
    """Extract discharge features for all cells for visualization"""
    features_list = []
    cycle_lives = []
    cell_ids = []

    feature_names = [
        "Minimum",
        "Variance",
        "Skewness",
        "Kurtosis",
        "Early discharge capacity",
        "Difference between max discharge capacity and early discharge capacity",
    ]

    print("Extracting discharge features for all cells...")
    # Get unique cell keys
    cell_keys = df["cell_key"].unique()

    for cell in cell_keys:
        cell_data = df[df["cell_key"] == cell]

        if "cycle_life" not in cell_data.columns or len(cell_data) == 0:
            continue

        cycle_life = cell_data["cycle_life"].iloc[0]

        features = get_discharge_features(cell_data)
        if features is not None:
            feature_vector = [features[name] for name in feature_names]
            features_list.append(feature_vector)
            cycle_lives.append(cycle_life)
            cell_ids.append(cell)

    print(f"Successfully extracted features for {len(features_list)} cells")
    return np.array(features_list), cycle_lives, cell_ids, feature_names


def visualize_pca(features, cycle_lives, cell_ids, feature_names):
    """Apply PCA to features and create a visualization colored by cycle life"""
    scaler = StandardScaler()
    features_scaled = scaler.fit_transform(features)

    pca = PCA(n_components=2)
    pca_result = pca.fit_transform(features_scaled)

    pca_df = pd.DataFrame({
        "PC1": pca_result[:, 0],
        "PC2": pca_result[:, 1],
        "cell_id": cell_ids,
        "cycle_life": cycle_lives,
    })

    plt.figure(figsize=(12, 10))
    scatter = plt.scatter(
        pca_df["PC1"], pca_df["PC2"], c=pca_df["cycle_life"], cmap="viridis", alpha=0.7, s=100
    )

    plt.colorbar(scatter, label="Cycle Life")
    plt.xlabel(f"Principal Component 1 ({pca.explained_variance_ratio_[0]:.2%} variance)")
    plt.ylabel(f"Principal Component 2 ({pca.explained_variance_ratio_[1]:.2%} variance)")
    plt.title("PCA of Discharge Features Colored by Cycle Life")
    plt.grid(True, linestyle="--", alpha=0.7)
    plt.tight_layout()

    print(f"Explained variance ratio: {pca.explained_variance_ratio_}")
    print(f"Total explained variance: {sum(pca.explained_variance_ratio_):.2%}")

    print("\nFeature loadings:")
    for i, feature in enumerate(feature_names):
        print(f"{feature}: PC1 = {pca.components_[0, i]:.4f}, PC2 = {pca.components_[1, i]:.4f}")

    return plt, pca_df


def visualize_tsne(features, cycle_lives, cell_ids, perplexity=30):
    """Apply t-SNE to features and create a visualization colored by cycle life"""
    scaler = StandardScaler()
    features_scaled = scaler.fit_transform(features)

    tsne = TSNE(n_components=2, perplexity=perplexity, random_state=42)
    tsne_result = tsne.fit_transform(features_scaled)

    tsne_df = pd.DataFrame({
        "TSNE1": tsne_result[:, 0],
        "TSNE2": tsne_result[:, 1],
        "cell_id": cell_ids,
        "cycle_life": cycle_lives,
    })

    plt.figure(figsize=(12, 10))
    scatter = plt.scatter(
        tsne_df["TSNE1"],
        tsne_df["TSNE2"],
        c=tsne_df["cycle_life"],
        cmap="viridis",
        alpha=0.7,
        s=100,
    )

    plt.colorbar(scatter, label="Cycle Life")
    plt.xlabel("t-SNE Component 1")
    plt.ylabel("t-SNE Component 2")
    plt.title(f"t-SNE of Discharge Features Colored by Cycle Life (perplexity={perplexity})")
    plt.grid(True, linestyle="--", alpha=0.7)
    plt.tight_layout()

    return plt, tsne_df


def run_discharge_visualizations():
    """Run both PCA and t-SNE visualizations on discharge features"""
    print("Loading datasets...")

    main_df = pd.read_csv("beep_main_ds.csv")
    secondary_df = pd.read_csv("test_secondary_beep_dataset.csv")

    df = pd.concat([main_df, secondary_df])
    print(f"Loaded data with {len(df)} rows from {df['cell_key'].nunique()} cells")
    features, cycle_lives, cell_ids, feature_names = extract_features_for_visualization(df)

    if len(features) == 0:
        print("No valid features extracted. Check the data and feature extraction.")
        return

    # Run PCA
    print("\n--- PCA Visualization of Discharge Features ---")
    pca_plt, pca_df = visualize_pca(features, cycle_lives, cell_ids, feature_names)
    pca_plt.savefig("discharge_pca.png")
    pca_plt.show()

    perplexity_values = [5, 30, 50]

    print("\n--- t-SNE Visualizations of Discharge Features ---")
    for perplexity in perplexity_values:
        print(f"\nRunning t-SNE with perplexity={perplexity}")
        tsne_plt, tsne_df = visualize_tsne(features, cycle_lives, cell_ids, perplexity)
        tsne_plt.savefig(f"discharge_tsne_perplexity_{perplexity}.png")
        tsne_plt.show()

    print("\nAll discharge feature visualizations completed!")


if __name__ == "__main__":
    run_discharge_visualizations()
```
